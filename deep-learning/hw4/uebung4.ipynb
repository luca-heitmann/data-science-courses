```python
a = torch.randn(1)
b = torch.randn(1)
d = torch.randn(1)

x = torch.randn(1, requires_grad=True)

def f(x):
  return a*x**6 + b*x**4 + d*torch.log(x)

output = f(x)

print(f"Output: {output.item()}")
print(f"Output shape: {output.shape}")

output.backward()

print(f"Gradient df/dx: {x.grad.item()}")
```

```python
x = torch.randn(1, requires_grad=True)

def f(x):
  return torch.e ** (-x) * torch.cos(-2 * x)

output = f(x)

print(f"Output: {output.item()}")
print(f"Output shape: {output.shape}")

output.backward()

print(f"Gradient df/dx: {x.grad.item()}")
```

```python
d = 3
x = torch.randn(d)
b = torch.randn(1)

w = torch.randn(d, requires_grad=True)

def g(z):
  return 1 / (1 + torch.e ** (-z))

def f(w):
  return g(w @ x + b)

output = f(w)

print(f"Output: {output.item()}")
print(f"Output shape: {output.shape}")

output.backward()

print(f"Gradient df/dw: {w.grad}")
```

```python
d = 3
k = 2
v = torch.randn(1, d)
a = torch.randn(k, 1)

X = torch.randn(d, k, requires_grad=True)

def f(X):
  return v @ X @ a

output = f(X)

print(f"Output: {output.item()}")
print(f"Output shape: {output.shape}")

output.backward()

print(f"Gradient df/dX: {w.grad}")
```

```python
d = 3
n = 2
z = torch.randn(d, 1)
v = torch.randn(1, d)
B = torch.randn(d, d)
A = torch.randn(d, d)

X = torch.randn(d, n, requires_grad=True)

def f(X):
  return v @ A @ X @ torch.t(X) @ B @ z

output = f(X)

print(f"Output: {output.item()}")
print(f"Output shape: {output.shape}")

output.backward()

print(f"Gradient df/dX: {w.grad}")
```
